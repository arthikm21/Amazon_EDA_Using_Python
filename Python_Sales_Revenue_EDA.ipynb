SCATTER PLOT

```python
x = df['Category']
y = df['Size']

plt.scatter(x,y)
plt.xlabel('Category')
plt.ylabel('Size')
plt.title('Scatter Plot')
plt.show()
```

**<span style= "font-size:30px;"> We can see have a clear view of the categories and the sizes that are sold with this diagram</span>**

```python
top_ten_states = df['Ship-State'].value_counts().head(10).sort_values(ascending=False)
subset_df = df[df['Ship-State'].isin(top_ten_states.index)]

plt.figure(figsize=(10, 8))
ax = sns.countplot(data=subset_df, x='Ship-State', order=top_ten_states.index)

for p in ax.patches:
    ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points', fontsize=8)

plt.xlabel('Ship-State')
plt.ylabel('Order Count')
plt.title('Distribution of State')
plt.xticks(rotation=90)
plt.show()
```

**<span style ="font-size:20px;">Buyers living in State of Maharashtra are the largest customer base for Amazon followed by Karnataka</span>**
